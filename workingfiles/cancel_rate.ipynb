```python
cancel_rate = pd.read_csv("D:\合作店铺\科帕奇\科帕奇_取消率.csv")
```

```python
cancel_rate.columns = ['Year','Month','Status','Market','Number']
```

```python
cancel_rate = cancel_rate.pivot(index=['Year','Month','Market'],columns='Status',values='Number').reset_index()
```

```python
cancel_rate = cancel_rate.fillna(0)
```

```python
cancel_rate['Total_Orders'] = cancel_rate['Canceled'] + cancel_rate['Pending'] + cancel_rate['Shipped'] + cancel_rate['Unshipped']
```

```python
cancel_rate
```

```python
cancel_rate['cancel_rate'] = cancel_rate['Canceled']/cancel_rate['Total_Orders']
```

```python
cancel_rate.to_csv('D:\合作店铺\科帕奇\Agg Data\科帕奇_订单取消率.csv')
```
